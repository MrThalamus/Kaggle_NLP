# **Remove HTML tags**

```python
#Remove HTML tags using regular expression
import re

def remove_html_tags(text):
    pattern = re.compile(r'<.*?>')
    return pattern.sub('', text)

```

```python
def chat_con(text):
    new_text = []
    for w in text.split():
        if w.upper() in slang_dict:
            new_text.append(slang_dict[w.upper()])
        else:
            new_text.append(w)
    return " ".join(new_text)
```

# **Redmoving stop words**

```python
from nltk.corpus import stopwords
#stopwords.words('spanish')
```

```python
def rem_stopwords(text):
    next_text = []
    for w in text.split():
        if w in stopwords.words('english'):
            next_text.append('')
        else:
            next_text.append(w)
    new = next_text[:]
    next_text.clear()
    return " ".join(new)
```

```python
rem_stopwords("Hello! I'm here for the session")
```

# **Emoji handling**

```python
# removing
import re

emoji_pattern = re.compile(
    "["
    "\U0001F600-\U0001F64F"  # emoticons
    "\U0001F300-\U0001F5FF"  # symbols & pictographs
    "\U0001F680-\U0001F6FF"  # transport & map
    "\U0001F1E0-\U0001F1FF"  # flags
    "\U00002700-\U000027BF"  # dingbats
    "\U0001F900-\U0001F9FF"  # supplemental symbols
    "\U00002600-\U000026FF"  # misc symbols
    "\U00002B00-\U00002BFF"  # arrows etc.
    "]+",
    flags=re.UNICODE,
)

def remove_emojis(text):
    return emoji_pattern.sub("", text)
```

```python
text = "I love NLP üòçüî• but cleaning text ü§ñ is important! üòÇ"
print(remove_emojis(text))
```
